```python
#import library
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

```

```python
df = pd.read_csv('/content/Education.csv')
df
```

```python
# Step 2: Convert the text data into numerical features using CountVectorizer (Bag of Words)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Text'])

# Step 3: Define the target variable (Labels)
y = df['Label']

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Initialize the Naive Bayes Classifier (MultinomialNB is commonly used for text classification)
nb_classifier = MultinomialNB()

# Step 6: Train the model
nb_classifier.fit(X_train, y_train)

# Step 7: Predict on the test set
y_pred = nb_classifier.predict(X_test)

# Step 8: Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Step 9: Test on new text input
new_text = ['The impact of educational reforms remains uncertain despite extensive research.']
new_text_transformed = vectorizer.transform(new_text)
prediction = nb_classifier.predict(new_text_transformed)

# Print the results
print(f"Predictions: {y_pred}")
print(f"Accuracy: {accuracy}")
print(f"Prediction for the new text: {prediction[0]}")
```
