```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Tạo dữ liệu giả định cho KNN
np.random.seed(42)
data_size = 1000
# Tạo các đặc trưng ngẫu nhiên giữa các lớp
X_class0 = np.random.multivariate_normal([2, 2], [[1.5, 0.75], [0.75, 1.5]], data_size // 2)
X_class1 = np.random.multivariate_normal([4, 4], [[1.5, 0.75], [0.75, 1.5]], data_size // 2)
X = np.vstack((X_class0, X_class1))
y = np.hstack((np.zeros(data_size // 2), np.ones(data_size // 2)))

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra với test_size=0.3 và random_state=42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Lưu các mẫu dữ liệu và nhãn tương ứng vào biến
sample_data = pd.DataFrame(X_train[:5], columns=["Feature 1", "Feature 2"])
sample_labels = y_train[:5]

# Chuyển sample_data thành dict để có thể JSON serialize được
result_3 = {
    "Sample_data": sample_data.to_dict(orient='records'),  # Chuyển DataFrame thành list of dicts
    "Sample_label": sample_labels.tolist()  # Chuyển labels thành list
}

# Hiển thị kết quả từ biến
print("Sample Data:\n", sample_data)
print("Corresponding Labels:", sample_labels)

# In kết quả dưới dạng dict
print(result_3)

```

```python
import numpy as np

# Hàm tính khoảng cách Euclidean giữa hai điểm
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

# Hàm KNN để dự đoán nhãn
def knn_predict(X_train, y_train, X_test, k=5):
    y_pred = []
    for test_point in X_test:
        # Tính khoảng cách từ điểm kiểm tra đến tất cả các điểm trong tập huấn luyện
        distances = [euclidean_distance(test_point, x) for x in X_train]
        # Lấy k chỉ số của các điểm gần nhất
        k_indices = np.argsort(distances)[:k]
        # Lấy nhãn của k điểm gần nhất
        k_nearest_labels = [y_train[i] for i in k_indices]
        # Chọn nhãn phổ biến nhất trong k nhãn gần nhất
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        y_pred.append(most_common)
    return np.array(y_pred)

# Dự đoán trên tập kiểm tra với k = 5
y_pred_knn = knn_predict(X_train, y_train, X_test, k=5)
```

```python
# Định nghĩa hàm confusion_matrix
def confusion_matrix(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    return np.array([[TN, FP], [FN, TP]])

# Hàm tính toán và in các chỉ số
def evaluate_model(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    TN, FP, FN, TP = cm.ravel()
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    print(f"Confusion Matrix:\n{cm}")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"F1 Score: {f1:.2f}")

# Đánh giá mô hình KNN
print("KNN Model Evaluation:")
result_4 = evaluate_model(y_test, y_pred_knn)
print(result_4)
```
